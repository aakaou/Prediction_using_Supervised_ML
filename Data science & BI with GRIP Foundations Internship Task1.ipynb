## Section 1

```python
# Importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
# link of data
url ='http://bit.ly/w-data'
```

```python
# Reading the data
data = pd.read_csv(url)
```

```python
# data imported successufully
data
```

```python
#  function is used to generate descriptive statistics of the dataset
data.describe()
```

```python
# Importing necessary libraries
import pandas as pd
from bokeh.layouts import gridplot
from bokeh.io import show, output_notebook
from bokeh.plotting import figure


output_notebook()
# Creating a figure for the histogram of Hours column
p1 = figure(title="Distribution of Study Hours", x_axis_label="Hours", y_axis_label="Frequency")
hist1, edges1 = np.histogram(data["Hours"], bins=10)
p1.quad(top=hist1, bottom=0, left=edges1[:-1], right=edges1[1:], fill_color="#036564", line_color="#033649")

# Creating a figure for the histogram of Scores column
p2 = figure(title="Distribution of Scores", x_axis_label="Scores", y_axis_label="Frequency")
hist2, edges2 = np.histogram(data["Scores"], bins=10)
p2.quad(top=hist2, bottom=0, left=edges2[:-1], right=edges2[1:], fill_color="#b30000", line_color="#800000")

# Combining the figures into a grid
grid = gridplot([[p1, p2]])


# Showing the plot
show(grid)
```


This part creates a scatter plot using Bokeh library to visualize the relationship between the study hours and the scores. 

```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import show, output_notebook
output_notebook()
plt.style.use('ggplot')
# Creating a figure object
p = figure(title="Scores vs Study Hours", x_axis_label="Study Hours", y_axis_label="Scores")

# Adding the scatter plot to the figure object
p.circle(data["Hours"], data["Scores"], size=10, color="blue")

# Showing the plot
show(p)
```

## Section 2

```python
# Extracting the features and target variable
X = data.iloc[:, :-1].values
y = data.iloc[:, 1].values
```

```python
# Splitting the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Creating the model and fitting it to the training data
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the test set results
y_pred = regressor.predict(X_test)
print(y_pred)
```

```python
# Visualizing the training set results
plt.style.use('ggplot')
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Study Hours vs Percentage (Training set)')
plt.xlabel('Study Hours')
plt.ylabel('Percentage')
plt.show()
```

```python
# Visualizing the test set results
plt.style.use('ggplot')
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Study Hours vs Percentage (Test set)')
plt.xlabel('Study Hours')
plt.ylabel('Percentage')
plt.show()
```

```python
# Making new prediction for 9.25 hrs/day
hours = [[9.25]]
new_prediction = regressor.predict(hours)
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(new_prediction[0]))
```

```python
# Use multiple hours to test your prediction.
hours = [[9.25], [8.5], [7.75]]  # A list of lists containing multiple hours
new_prediction = regressor.predict(hours)
for i in range(len(hours)):
    print("No of Hours = {}".format(hours[i]))
    print("Predicted Score = {}".format(new_prediction[i]))
```
